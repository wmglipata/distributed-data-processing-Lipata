**Is the data processed sequentially or distributed?**
- The data was processed in a distributed manner.

**What does Spark do internally?**
- Spark acts as the distributor for worker cores when reading the CSV. It slices the data into partitions and hands a small stack of rows to each of the workers to read at the same time, otherwise known as Distributed Processing. Beyond just splitting the work, Spark also uses Lazy Evaluation, meaning it doesn't actually start moving the data until an action is called like .show(). This allows it to build an optimized plan(Directed Acyclic Graph) first. Once it starts, Spark performs these calculations In-Memory (RAM) rather than constantly writing to the slow hard drive, which is why it can process thousands of rows in the time it takes standard Python to read just a few.
